## Customized Model

The following cell define the customized model for this multi-label classification task.

This model contains:

resnet18 - for image feature extraction

TinyBERT - for caption feature extraction

```python
class MLCModel(nn.Module):
    def __init__(self, num_classes):
        super(MLCModel, self).__init__()
        # load pretrained resnet18 for image feature extraction
        resnet18 = models.resnet18(pretrained=True)
        # load pretrained TinyBERT for caption feature extraction
        tiny_BERT = AutoModelForSequenceClassification.from_pretrained("cross-encoder/ms-marco-TinyBERT-L-2-v2")

        self.image_model = resnet18.to(device)
        self.nlp_model = tiny_BERT.to(device)

        # modify the fc layer of resnet
        image_num_features = self.image_model.fc.in_features
        self.image_model.fc = nn.Linear(image_num_features, image_num_features)

        # The number of features of the last hidden layer of tiny_BERT
        nlp_num_features = 128

        # The classifier for final classification output
        self.classifier = nn.Sequential(
            nn.Linear(image_num_features + nlp_num_features, image_num_features),
            nn.ReLU(),
            nn.Dropout(p=0.5),
            nn.Linear(image_num_features, 256),
            nn.ReLU(),
            nn.Dropout(p=0.5),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(p=0.5),
            nn.Linear(128, num_classes),
            # nn.Sigmoid()
        )

    def forward(self, image, caption):

        # extract image feature
        image_feature = self.image_model(image)

        # extract caption feature
        caption['input_ids'] = caption['input_ids'].to(device)
        caption['attention_mask'] = caption['attention_mask'].to(device)
        caption_feature = self.nlp_model(**caption, output_hidden_states=True).hidden_states[-1].mean(dim=1)

        # concatinate image and caption features
        feature = torch.cat([image_feature, caption_feature], dim=-1)

        # send features to classifier then produce output
        output = self.classifier(feature)

        return output
```

```python
# load tokenizer for pretrained TinyBERT
tokenizer = AutoTokenizer.from_pretrained("cross-encoder/ms-marco-TinyBERT-L-2-v2")
```

```python
# find the max caption lenght
with open(TRAIN_CSV) as file:
    lines = [re.sub(r'([^,])"(\s*[^\n])', r'\1/"\2', line) for line in file]
    train_data = pd.read_csv(StringIO(''.join(lines)), escapechar="/")

with open(TEST_CSV) as file:
    lines = [re.sub(r'([^,])"(\s*[^\n])', r'\1/"\2', line) for line in file]
    test_data = pd.read_csv(StringIO(''.join(lines)), escapechar="/")

max_caption_length = 0
for caption in train_data['Caption']:
    tokenized_caption = tokenizer(caption)
    length = len(tokenized_caption['input_ids'])
    if length > max_caption_length:
        max_caption_length = length

for caption in test_data['Caption']:
    tokenized_caption = tokenizer(caption)
    length = len(tokenized_caption['input_ids'])
    if length > max_caption_length:
        max_caption_length = length

print("Max caption length:", max_caption_length)
```

## Customize Data
Customizing the both image and caption data for dataloader

```python
class CustomizeData(Dataset):
    def __init__(self, annotations_file, img_dir, transform=None, mode='train'):
        """
        Customize the dataset for DataLoader

        Params:
            annotations_file: path to the csv train/test file
            img_dir: image directory
            transform: optional transform
            mode: set to 'train' or 'test'

        Returns:
            three element tuple:
              image, label, caption or
              image, img_name, caption
        """
        self.label_file = annotations_file
        self.img_dir = img_dir
        self.transform = transform
        self.num_classes = 20
        self.mode = mode

        with open(self.label_file) as label_file:
            lines = [re.sub(r'([^,])"(\s*[^\n])', r'\1/"\2', line) for line in label_file]
            self.img_labels = pd.read_csv(StringIO(''.join(lines)), escapechar="/")

    def __len__(self):
        return self.img_labels.shape[0]

    def __getitem__(self, idx):
        # load and transform image
        img_name = self.img_labels.iloc[idx, 0]
        img_path = os.path.join(self.img_dir, self.img_labels.iloc[idx, 0])
        image = read_image(img_path).to(device)
        if self.transform:
            image = self.transform(image)

        if self.mode == 'train':
            # load and tokenize caption
            caption = self.img_labels.iloc[idx,2]
            caption = tokenizer(caption, return_tensors='pt', padding='max_length', max_length=max_caption_length)
            # load and parse label
            label = self.img_labels.iloc[idx,1]
            label = self.parse_labels(label)
            label = torch.tensor(label, dtype=torch.float32)
            return image, label, caption
        else:
            # for testing, we don't need label
            caption = self.img_labels.iloc[idx,1]
            caption = tokenizer(caption, return_tensors='pt', padding='max_length', max_length=max_caption_length)
            return image, img_name, caption

    '''
    There are 18 labels from 1-19 missing 12,
    we use this special way to parse labels and transfer to one-hot code
    '''
    def parse_labels(self, label_str):
        label_indices = list(map(int, label_str.split()))
        label_vector = [0]*self.num_classes
        for i in label_indices:
            label_vector[i] = 1
        del label_vector[12]
        del label_vector[0]
        return label_vector
```

#### Customize collate_fn for DataLoader

```python
def collate_fn(batch):
    images, labels, captions = zip(*batch)

    images = torch.stack(images)

    # If training, labels are tensors, we need to stack them together
    # If testing, labels are str which is the image name, do nothing
    if isinstance(labels[0], torch.Tensor):
        labels = torch.stack(labels)

    # stack input_ids and attention_mask of tokenized caption
    # into shape: (batch_size, sequence_length)
    input_ids = torch.stack([caption['input_ids'].squeeze(0) for caption in captions])
    attention_mask = torch.stack([caption['attention_mask'].squeeze(0) for caption in captions])

    # recombine input_ids and attention_mask into dictonary
    captions = {'input_ids': input_ids, 'attention_mask': attention_mask}

    return images, labels, captions
```

## Data Loading and Data Preprocessing

```python
# Transform and augment images
transforms = {
    'train': v2.Compose([
        # v2.RandomResizedCrop(size=(224,224), antialias=True),
        v2.Resize((224, 224)),
        v2.RandomHorizontalFlip(),
        v2.ColorJitter(),
        # v2.Resize(256),
        # v2.CenterCrop(224),
        # v2.ToImage(),
        v2.RandomAffine(degrees=20,
                        translate=(0.2, 0.2),
                        scale=(0.5, 1.5),
                        fill=tuple(np.array(np.array([0.485, 0.456, 0.406]) * 255).astype(int).tolist())),
        v2.ToDtype(torch.float32, scale=True), # Transfer images to tensor and scale to [0,1]
        v2.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ]),
    'test': v2.Compose([
        v2.Resize((224,224)),
        # v2.CenterCrop(224),
        # v2.ToImage(),
        v2.ToDtype(torch.float32, scale=True), # Transfer images to tensor and scale to [0,1]
        v2.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
}

# set random seed
torch.manual_seed(42)

dataset = CustomizeData(annotations_file=TRAIN_CSV, img_dir=IMAGE_DIR, transform=transforms['train'], mode='train')
test_dataset = CustomizeData(annotations_file=TEST_CSV, img_dir=IMAGE_DIR, transform=transforms['test'], mode= 'test')
train_dataset, val_dataset = torch.utils.data.random_split(dataset,
                                                           [int(TRAIN_VAL_PROP * len(dataset)),
                                                            len(dataset) - (int(TRAIN_VAL_PROP * len(dataset)))])

train_loader = DataLoader(dataset=train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=0, collate_fn=collate_fn)
val_loader = DataLoader(dataset=val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=0, collate_fn=collate_fn)
test_loader = DataLoader(dataset=test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=0, collate_fn=collate_fn)
```

## Training and Evaluation

#### Define f1_score calculation method

```python
def f1_calculation(outputs, labels):
    """
    Calculate the f1 score for predictions and labels.

    Params:
      outputs (torch.Tensor): the output of the model.
      labels (torch.Tensor): the actual labels.

    Returns: f1 score
    """
    t_list = [0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9]

    best_f1 = 0
    for threshold in t_list:
        predicted = (torch.sigmoid(outputs) > threshold)
        # predicted = torch.where(outputs > threshold, torch.ones_like(outputs), torch.zeros_like(outputs)).int()
        labels = labels.int()

        tp = (predicted & labels).sum().float()
        fp = (predicted & (~labels)).sum().float()
        fn = ((~predicted) & labels).sum().float()

        p = tp / (tp + fp)
        r = tp / (tp + fn)
        f1 = 2 * ((p * r) / (p + r))

        if f1 > best_f1:
            best_f1 = f1

    return best_f1
```

#### Define training method

```python
def train(model, criterion, optimizer, max_epochs):

    train_loss_list = []
    train_f1_list = []
    val_loss_list = []
    val_f1_list = []

    pre_f1_score = 0

    for epoch in range(max_epochs):

        epoch_start_time = time()
        train_loss = 0.0
        val_loss = 0.0
        train_f1_scores = 0.0
        val_f1_scores = 0.0

        # set model to train mode to record loss and f1_score on training set
        model.train()
        train_count = 0
        for images, labels, captions in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            # forward pass
            outputs = model(images, captions)
            loss = criterion(outputs, labels)
            # backward and optimize
            loss.backward()
            optimizer.step()

            # record the loss of current batch
            train_loss += loss.item() * images.size(0)
            # record the train f1 score of current batch
            train_f1_scores += f1_calculation(outputs, labels)
            train_count += 1

        # set model to evaluation mode, to record loss and f1_score of validaiton set
        model.eval()
        val_count = 0
        with torch.no_grad():
            for images, labels, captions in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images, captions)

                # val loss
                loss = criterion(outputs, labels)
                val_loss += loss.item() * images.size(0)
                # record the val f1 score of current batch
                val_f1_scores += f1_calculation(outputs, labels)
                val_count += 1

        # calculate train loss and f1 score of current epoch
        epoch_train_loss = train_loss / len(train_loader.dataset)
        train_f1 = 100 * (train_f1_scores / train_count)
        train_loss_list.append(epoch_train_loss)
        train_f1_list.append(train_f1)

        # calculate train loss and f1 score of current epoch
        epoch_val_loss = val_loss / len(val_loader.dataset)
        val_f1 = 100 * (val_f1_scores / val_count)
        val_loss_list.append(epoch_val_loss)
        val_f1_list.append(val_f1)

        epoch_runtime = time() - epoch_start_time

        print(f'Epoch {epoch+1}/{max_epochs} used: {epoch_runtime:.2f} seconds')
        print(f'Train: Loss: {epoch_train_loss:.4f}, F1 Score: {train_f1:.2f}')
        print(f'Validation: Loss: {epoch_val_loss:.4f}, F1 Score: {val_f1:.2f}')
        print('------------------------------------------------')

        if val_f1 > pre_f1_score:
            pre_f1_score = val_f1
        if pre_f1_score > val_f1 and pre_f1_score >= 88.5:
            print('Training early stoped')
            break

    result_dict = {
        'train_loss': train_loss_list,
        'val_loss': val_loss_list,
        'train_f1': train_f1_list,
        'val_f1': val_f1_list
    }
    return result_dict
```

## Test the best threshold

```python
t_list = [0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9]
best_threshold = 0
best_score = 0
for i in t_list:
    def f1_calculation(outputs, labels):
        predicted = torch.sigmoid(outputs) > i
        # predicted = torch.where(outputs > i, torch.ones_like(outputs), torch.zeros_like(outputs)).int()
        labels = labels.int()

        tp = (predicted & labels).sum().float()
        fp = (predicted & (~labels)).sum().float()
        fn = ((~predicted) & labels).sum().float()

        p = tp / (tp + fp)
        r = tp / (tp + fn)
        f1 = 2 * ((p * r) / (p + r))

        return f1

    model.eval()
    with torch.no_grad():
        scores = []
        for images, labels, captions in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images, captions)
            scores.append(f1_calculation(outputs, labels))
        f1 = (100 * sum(scores) / len(scores))
        if f1 > best_score:
            best_score = f1
            best_threshold = i
        print(f'{i} ---- F1 score: {(100 * sum(scores) / len(scores)):.2f}')
print(best_threshold)
print(best_score)
```

## Output prediction

```python
def map_index_to_label(index):
    if index >= 11:  # missing 0 and 12, need to add 2
        return index + 2
    else:
        return index + 1 # others only missing 0, so add 1

def predict_test_set(model, dataloader):
    model.eval()
    predictions = []
    image_ids = []

    with torch.no_grad():
        for images, ids, captions in dataloader:
            images = images.to(device)
            outputs = model(images, captions)
            predicted_probs = torch.sigmoid(outputs)  # using sigmoid function to transfer the prediction into range[0,1]
            predicted_labels = (predicted_probs > best_threshold).int()  # using threshold to convert prediction to classes
            for pred, id_ in zip(predicted_labels.cpu().numpy(), ids):
                # map index of labels to actural class
                mapped_labels = [map_index_to_label(i) for i in np.where(pred == 1)[0]]
                predictions.append(' '.join(map(str, mapped_labels)))
                image_ids.append(id_)

    # save the predition
    with open("test_predictions.csv", "w") as f:
        f.write("ImageID,Labels\n")
        for id_, pred in zip(image_ids, predictions):
            f.write(f"{id_},{pred}\n")
    print("Predictions have been saved to 'test_predictions.csv'.")

# predict on test set
predict_test_set(model, test_loader)
```
